```python
import matplotlib.pyplot as plt
```

```python
plt.boxplot(df_automotora["id_modelo"],patch_artist = True,boxprops = dict(facecolor = "lightblue"),labels=['Modelo'],)
plt.title("Modelos De automoviles ")
plt.xlabel('Id_Auto')
plt.ylabel('Marca de los  auto ')
```

 Grafico Boxplot: Se puede analisar que los modelos de auto que tiene una poliza de seguros van desde  800 hasta los 3100 con ese dato se puede realizar difrenete analisis uno de ellos.son los dueño de autos que tiene a compran una frecuentemente una  poliza.


```python
plt.hist(df_automotora["id_cobertura"])
plt.title("Histograma Cobertura de poliza ")
plt.xlabel('Numero de Cobertura ')
plt.ylabel('Cobertura ')
plt.show()

```

Se puede analizar que las cobertura de seguro mas frecuente esta en 4 y 7 segun lo repreenta el grafico

```python
plt.scatter(df_automotora["id_marca"],df_automotora["id_sexo"],alpha=0.3,c="red")
plt.title('Grafico de Punto ')
plt.xlabel('Marca de Automovil ')
plt.ylabel('Sexo ')
```
